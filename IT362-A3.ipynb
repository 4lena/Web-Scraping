## Task 1:

#### Category 1: Trolleys

```python
Trolleys = 'https://www.ikea.com/sa/en/cat/trolleys-fu005/'
Names = []
Price = []
Description = []
Category = []

try:
    html = requests.get(Trolleys)
    soup = BeautifulSoup(html.text, 'html')
    #print(soup)
    
    for i in soup.find_all(class_='pip-header-section__title--small notranslate'):
        Names.append(i.text)
        
    for i in soup.find_all(class_='pip-price__integer'):
        Price.append(i.text)
        
    for i in soup.find_all(class_='pip-header-section__description-text'):
        Description.append(i.text)
    
    Items_len = len(Names)
    
    for i in range(Items_len):
        Category.append("Trolleys")
        
    time.sleep(4)
    print ("Scraping has been completed successfully!")
               
except:
    print ("There was something wrong!")
```
